```python
# SEABORN
# It is mainly use for complex data.

## Matplotlib                             Seaborn
# 1) Plot simple data                1) Plot complex data
# 2) Limited plots                   2) more number of plots
# 3) Limited attributes              3) More no. of attributes

# Types of plots in Seaborn
# 1) Distribution  Plot
# 2) Categorical Plot
# 3) Matrix plot
# 4) Regression plot.

# Based on number of column
# 1) Univariate plot :- It plots the graph based on one column
# 2) Bivariate plot :- It plots the graph based on two column
# 3) Multi-variate plot :- It plots the graph more than 2 column.

```

# Mtrixplot():
- It is used to compare one column with respect to another column.
- It is used mainly used in Machine learning.
- It is used to observe the relation between multiple columns.
# Syntax:
- columns('col1','col2'........'coln')
- sns.heatmap(data=var[column].corr(),annote=True,cmap'colwarm')

```python
tips
```

```python
col = ['tip','size']
sns.heatmap(data=tips[col].corr())
```

```python
col = ['tip','size']                                       # annot - showing number
sns.heatmap(data=tips[col].corr(),annot=True,cmap='inferno')                
```

```python
hr
```

```python
col = ['number_project','promotion_last_5years','left','number_project']                                     
sns.heatmap(data=hr[col].corr(),annot=True,cmap='inferno') 
```
