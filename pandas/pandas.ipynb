```python
df = pd.DataFrame({
    'Name': ['A', 'B', 'C'],
    'Marks': [90, 85, 80]
})

df['Result'] = ['Pass', 'Pass', 'Pass']

print(df)
```

```python
df['School'] = 'K.N.I.P.S.S'
```

```python
df['Status'] = df['Marks'].apply(lambda x: 'Pass' if x >= 80 else 'Fail')
```

```python
df['Double_Marks'] = df['Marks'] * 2
```

```python
df.insert(1, 'Age', [18, 19, 20])   # position 1 me column add
```

```python
df = df.assign(Grade=['A', 'B', 'C'])
```

```python
df['Category'] = ['Topper' if x > 85 else 'Average' for x in df['Marks']]
```

```python
df
```
