# Group By: group by() is used when we want to split the data into groups on a column.

# Use: If we want to do operation based on particular value,we go with group by 

# Syntax: 

# var= df.groupby('categorical_column')
# var.groups
# var.get_group('group_name')

```python
df = pd.read_csv('https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/hr_data.csv')
df.head()
```

```python
df.info()
```

```python
df['department'].unique()
```

```python
dept = df.groupby('department')
dept.groups
```

```python
gr = dept.get_group('IT')
gr
```

```python
sales = dept.get_group('sales')
sales
```

```python
df['salary'].unique()
```

```python
sal = df.groupby('salary')
sal.groups
```

```python
l=sal.get_group("low")
m=sal.get_group("medium")
n=sal.get_group("high")
```
