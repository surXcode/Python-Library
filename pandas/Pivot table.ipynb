# Joins - 
- If we want to fetch the data from to diffrent table s
- 1. inner join
  2. left join
  3. right join
  4. outer join

* inner join:-  it return only matching records from both the tables.

* outer join:-  It will returns all the record from both data frame.

# syntax:-
  pd.merge(var1,var2,on='common_col',how='inner/left')

# Feature engineering:-  It is a process of create a new column based on existing column.

* syntax :-
*    var[col]=new_valve

```python
df1 = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/titanic.csv")
df1
```

```python
df1['family_count_without'] = df1['sibsp']+df1['parch']
df1['family_count_with'] = df1['sibsp']+df1['parch']+1
df1
```

```python
out = []
for i in df1['name']:
        out.append(i.split(',')[0])
df1['family_name_for'] = out
df1
```

```python
df1['title'] = df1['name'].str.split('.').str[0].str.split(',').str[1]
df1
```
