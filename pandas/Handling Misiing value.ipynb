```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'name':['Suraj','Anuj','Ankit','Ashvik','Suyash',np.nan],
                 'roll_no':[np.nan,11,12,np.nan,14,16],
                 'marks':[58,44,np.nan,98,33,55]})
df
```

```python
df.dropna(inplace=True)
df
```

# Filling Nan Values:

# Syntax:
# var['col_name']=var['col_name'].fillna(value)

```python
df = pd.DataFrame({'name':['Suraj','Anuj','Ankit','Ashvik','Suyash',np.nan],
                 'roll_no':[np.nan,11,12,np.nan,14,16],
                 'marks':[58,44,np.nan,98,33,55]})
df
```

```python
df['name']=df['name'].fillna('unknown')
df
```

```python
m= round(df['marks'].mean())
df['marks'] = df['marks'].fillna(m).astype(int)
df
```

```python
df['roll_no'].mean()
```

```python
round(df['roll_no'].mean())
```

```python
m= round(df['roll_no'].mean())
df['roll_no'] = df['roll_no'].fillna(m).astype(int)
df
```

```python
m= round(df['roll_no'].mean())
df['roll_no'] = df['roll_no'].fillna(m).astype(bool)
df
```

```python
m= round(df['marks'].mean())
df['marks'] = df['marks'].fillna(m).astype(bool)
df
```

```python
df = pd.DataFrame({'name':['Suraj','Anuj','Amit','Ashvik','Suyash',np.nan],
                 'roll_no':[np.nan,11,12,np.nan,14,16],
                 'marks':[58,44,np.nan,98,33,55]})
df
```

# Forward filling:

- It will fill the nan value with its previous value.

# Syntax:
- df['col_name'] = df['col_name'].ffill()

```python
df['marks']=df['marks'].ffill()
df
```

# Backward filling:
- It will fill the nan value with its next value.
# Syntax:
- df['col_name'] = df['col_name'].bfill()

```python
df['marks']=df['marks'].bfill()
df
```

```python
df = pd.DataFrame({'name':['Suraj','Anuj','Amit','Ashvik','Suyash',np.nan],
                 'roll_no':[np.nan,11,12,np.nan,14,16],
                 'marks':[np.nan,44,np.nan,98,33,55]})
df
```

```python
df['marks']=df['marks'].bfill().ffill()
df
```
