# Boolean Factorial:




```python
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/titanic.csv")
df
```

# Boolean values:

# Syntax: 

- df[df['col_name']==value]

```python
df[df['sex']=='male']
```

```python
df[df['survived']==1]
```

```python
df[df['survived']==0]
```

```python
# Fetch the passenger data where the age is more than 50.

df[df['age']>50]
```

```python
# Fetch male passenger data who are survived.
df[(df['sex']=='male') & (df['survived']==1)]
```

```python
df[(df['pclass']==2) | (df['pclass']==3)]
```

```python
df.describe(include=object)
```

```python
df.drop_duplicates(inplace=True)
```
