```python
import numpy as np
```

```python
# 1Ô∏è‚É£ Create array
marks = np.array([45, 67, 89, 23, 56, 90])
print("Marks:", marks)
```

```python
# 2Ô∏è‚É£ Array properties
print("Shape:", marks.shape)
print("Dimensions:", marks.ndim)
print("Size:", marks.size)
print("Data type:", marks.dtype)
```

```python
# 3Ô∏è‚É£ Mathematical operations
print("Add grace marks:", marks + 5)
print("Square of marks:", marks ** 2)
print("Square root:", np.sqrt(marks))
```

```python
# 4Ô∏è‚É£ Statistical functions
print("Total:", np.sum(marks))
print("Average:", np.mean(marks))
print("Max:", np.max(marks))
print("Min:", np.min(marks))
print("Std Dev:", np.std(marks))
```

```python
# 5Ô∏è‚É£ Filtering (Boolean masking)
passed = marks[marks >= 40]
print("Passed students:", passed)
```

```python
# 6Ô∏è‚É£ Sorting
print("Sorted marks:", np.sort(marks))
```

```python
# 7Ô∏è‚É£ Unique values
print("Unique marks:", np.unique(marks))
```

```python
# 8Ô∏è‚É£ Where condition
result = np.where(marks >= 40, "Pass", "Fail")
print("Result:", result)
```

```python
# 9Ô∏è‚É£ Reshaping
matrix = marks.reshape(2, 3)
print("Reshaped matrix:\n", matrix)
```

```python
# üîü Transpose
print("Transpose:\n", matrix.T)
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ Flattening
print("Flatten:", matrix.flatten())
```

```python
# 1Ô∏è‚É£2Ô∏è‚É£ Random numbers (bonus marks)
bonus = np.random.randint(1, 6, size=6)
print("Bonus marks:", bonus)

final_marks = marks + bonus
print("Final marks:", final_marks)
```

```python

# 1Ô∏è‚É£3Ô∏è‚É£ Broadcasting
print("Increase all marks by 10:", final_marks + 10)
```

```python
# 1Ô∏è‚É£4Ô∏è‚É£ Linear algebra
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print("Matrix multiplication:\n", np.dot(a, b))
```

```python

# 1Ô∏è‚É£5Ô∏è‚É£ Identity matrix
print("Identity matrix:\n", np.eye(3))
```

```python
# 1Ô∏è‚É£6Ô∏è‚É£ Clipping values
print("Clipped marks (0-100):", np.clip(final_marks, 0, 100))
```
