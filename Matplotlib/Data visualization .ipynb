# Annotate:

```python
plt.plot(x,y,marker='s',markersize=10,markerfacecolor='#05DF72',linewidth=3,color='blue',linestyle=':')
plt.annotate("Center", xy=(5, 25))
```

```python
plt.plot(x,y,marker='*',markersize=10,markerfacecolor='#05DF72',linewidth=3,color='blue',linestyle='-')
plt.ylabel('sqr numbers',color='blue')
plt.xlabel('numbers',color='brown')
plt.title('Values',color='red',fontsize=20)
plt.legend('square of values')
for i in range(len(x)):
    plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]))
```

```python
plt.plot(x,y,marker='*',markersize=10,markerfacecolor='#05DF72',linewidth=3,color='blue',linestyle='-')
plt.ylabel('sqr numbers',color='blue')
plt.xlabel('numbers',color='brown')
plt.title('Values',color='red',fontsize=20)
plt.legend('square of values')
for i in range(len(x)):
    plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]),va='top')    # VA - vertical alingment
```

```python
plt.plot(x,y,marker='*',markersize=10,markerfacecolor='#05DF72',linewidth=3,color='blue',linestyle='-')
for i in range(len(x)):
     plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]),va='top',ha='left')    # HA - horizental alingment
```

```python
plt.plot(x,y,marker='*',markersize=10,markerfacecolor='#05DF72',linewidth=3,color='blue',linestyle='-')
for i in range(len(x)):
     plt.annotate(f'({x[i]},{y[i]})',(x[i],y[i]),va='top',ha='right')    # HA - horizental alingment
```

```python
z = x**2
plt.plot(x, z)

for i in range(len(x)):
    plt.annotate(f'({x[i]},{z[i]})', (x[i], z[i]))
```

```python
for i in range(len(x)):
    plt.annotate(f'({x[i]},{y[i]})',
                 xy=(x[i], y[i]),
                 xytext=(10, 10),          # text shift
                 textcoords='offset points',
                 arrowprops=dict(arrowstyle='->'))
```
