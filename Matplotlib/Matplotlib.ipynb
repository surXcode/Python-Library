# Pie chart

# pie chart:

- It is used to represent data in the form of percentage.

# Syntax:

 - plt.pie(data,label=[],autopct='%0,1f%%,color=[],shadow=True,explode=())

```python
viwers = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Danger']
plt.pie(viwers,labels=l)
```

```python
viwers = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Danger']
plt.pie(viwers,labels=l)
plt.show()
```

```python
viwers = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Danger']
plt.pie(viwers,labels=l,colors=['green','red','black','yellow','orange'])
plt.show()
```

```python
viwers = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Danger']
plt.pie(viwers,labels=l,colors=['green','red','brown','yellow','orange'],autopct='%0.2f%%',shadow=True,)
plt.show()
```

```python
viwers = [1500,5000,2000,1000,2500]
l = ['Drama','Action','Horror','Comedy','Danger']
plt.title('Movies')
plt.pie(viwers,labels=l,colors=['green','red','brown','yellow','orange'],autopct='%0.2f%%',shadow=True,explode=(0.1,0.0,0.1,0.1,0.1))
plt.show()
```
